```python
import pandas as pd

# Prepare a summary table for Table 1: Sampling sites, sample counts, climate zones, and buffer size
def summarize_sampling_sites_with_buffer(sites):
    summary = []
    for site in sites:
        df = site['df']
        area = df['Area'].iloc[0]
        biome = df['Biome'].iloc[0]
        buffer_km = site.get('buffer_km', 'N/A')
        counts = df['status'].value_counts().to_dict()
        protected_count = counts.get('protected', 0)
        unprotected_count = counts.get('unprotected', 0)
        total = len(df)
        summary.append({
            'Sampling Site': area,
            'Climate Zone': biome,
            'Buffer Zone (km)': buffer_km,
            'Protected Samples': protected_count,
            'Unprotected Samples': unprotected_count,
            'Total Samples': total
        })
    return pd.DataFrame(summary)

# List your main DataFrames for each site (IBI version) and their buffer sizes (as used in Area_Data or Area_Data_Buffer)
sampling_sites = [
    {'df': WetuDRC_Combined_Outside, 'buffer_km': 10},  # Custom outside, not a buffer
    {'df': KakamegaForests_Combined, 'buffer_km': 10},  # Custom outside, not a buffer
    {'df': PendjariNP_combined, 'buffer_km': 10},       # Custom outside, not a buffer
    {'df': KrugerNP_outside_combined, 'buffer_km': 10}, # Custom outside, not a buffer
    {'df': NiokoloNPSenegal_outside_combined, 'buffer_km': 10}, # Custom outside, not a buffer
    {'df': Namibia_Outside_combined, 'buffer_km': 10},  # Custom outside, not a buffer
    {'df': Multiple_Protected_Areas_Humid_Combined, 'buffer_km': 10}, # Custom outside, not a buffer
    {'df': Multiple_protected_Areas_SemiHumid_combined, 'buffer_km': 10}, # Custom outside, not a buffer
    {'df': Multiple_Unprotected_Areas_Humid_1_combined, 'buffer_km': 10}, # Custom outside, not a buffer
]

# If you know the actual buffer distances (in km) used for each, update the 'buffer_km' values accordingly.
# For example, if WetuDRC_Combined_Outside used a 30 km buffer, set 'buffer_km': 30

sampling_table = summarize_sampling_sites_with_buffer(sampling_sites)
sampling_table
```
